# 1. Import libraries


```python
import pandas as pd
```

# 2. Load data


```python
# 2. Load data
df = pd.read_csv('tests/raw_data.csv')
df
```

# 3. Process data


```python
targeted_languages = ["English", "Vietnamese"]
df_expanded = pd.DataFrame([
    {'japanese_paragraph': row['japanese_paragraph'], 'targeted_language': lang}
    for _, row in df.iterrows()
    for lang in targeted_languages
])
df_expanded
```

# 4. Save data


```python
df_expanded.to_csv('tests/test_cases.csv', index=False)
```
